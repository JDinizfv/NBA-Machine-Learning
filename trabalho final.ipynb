## 1.2 O que é a NBA?

A National Basketball Association (em português: Associação Nacional de Basquetebol; abreviação oficial: NBA) é a principal liga de basquetebol profissional da América do Norte. Com 30 franquias (29 nos Estados Unidos e 1 no Canadá), a NBA também é considerada a principal liga de basquete do mundo. É um membro ativo da USA Basketball (USAB),[2] que é reconhecida pela FIBA (a Federação Internacional de Basquetebol) como a entidade máxima e organizadora do basquetebol nos Estados Unidos. A NBA é uma das 4 'major leagues' de esporte profissional na América do Norte. Os jogadores da NBA são os atletas mais bem pagos do mundo, por salário médio anual.[3]

A liga foi fundada na cidade de Nova Iorque em 6 de Junho de 1946, como a Basketball Association of America (BAA).[4] A liga adotou o nome de National Basketball Association em 1949 quando se fundiu com a rival National Basketball League (NBL). A liga tem diversos escritórios ao redor do mundo, além de vários dos próprios clubes fora da sede principal na Olympic Tower localizada na Quinta Avenida 645. Os estúdios da NBA Entertainment e da NBA TV são localizados em Secaucus, Nova Jérsia.

Atualmente a liga é transmitida em Portugal pela Sport TV e pela NBA TV, e no Brasil pela Band na TV aberta, e pela ESPN e SporTV na TV fechada, além de transmissões em três canais do YouTube (NBA Brasil, Budweiser Brasil e TNT Sports Brasil) e na Twitch (Gaules). Nos Estados Unidos, as emissoras são NBA TV, TNT e ESPN, além da transmissão dos principais jogos pela ABC. O pacote NBA League Pass, que permite ao assinante assistir todos os jogos da liga ao vivo e em alta definição, é disponibilizado para assinatura em todo o mundo.

# 4 Pipelines de Transformação

```python
#Criando um pipeline
colunas_cat = ['year','name','playerId','playerSlug','positionId','teamId','status','position','college','draftinfo','birthplace','team','experience','birthdate','htwt']

NBAstats = strat_train_set.drop('salary',axis=1) # X without 1's
NBAstats_labels = strat_train_set['salary'].copy()
NBAstats_num = NBAstats.drop([col for col in colunas_cat if col in NBAstats.columns],axis=1) # numerical columns only
NBAstats_cat = NBAstats[[col for col in colunas_cat if col in NBAstats.columns]] # Categorical attributes only
```

```python
# Colunas numéricas
numeric_cols = NBAstats.select_dtypes(include=[np.number]).columns.tolist()

print("Colunas Numéricas: ", numeric_cols)

# Colunas categóricas
categorical_cols = NBAstats.select_dtypes(include=['category']).columns.tolist()

print("Colunas Categóricas: ", categorical_cols)
```

## 4.3 Pipeline para atributos numéricos e categóricos


```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self # Nothing else to do

    def transform(self, X, y=None):
        fieldGoalsRel = np.divide(X[:, 31], X[:, 32], out=np.zeros_like(X[:, 31]), where=X[:, 32]!=0)
        threePointFieldGoalsRel = np.divide(X[:, 33], X[:, 34], out=np.zeros_like(X[:, 33]), where=X[:, 34]!=0)
        freeThrowsRel = np.divide(X[:, 35], X[:, 36], out=np.zeros_like(X[:, 35]), where=X[:, 36]!=0)

        return np.c_[X, fieldGoalsRel, threePointFieldGoalsRel, freeThrowsRel]

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

num_attribs = NBAstats.select_dtypes(include=[np.number]).columns.tolist()

cat_attribs = NBAstats.select_dtypes(include=['category']).columns.tolist()

num_pipeline = Pipeline([('imputer', SimpleImputer(strategy="mean")),
                         ('adder', CombinedAttributesAdder()),
                         ('scaler', StandardScaler())])

cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy='constant',fill_value='missing')),
                         ('encoder', OneHotEncoder(sparse=False))])

full_pipeline = ColumnTransformer(transformers = [('num_pipeline', num_pipeline, num_attribs),
                                                  ('cat_pipeline', cat_pipeline, cat_attribs)])

NBAstats_prepared = full_pipeline.fit_transform(NBAstats)
print(NBAstats_prepared.shape)
NBAstats_prepared
```

Por definição um jogador superfaturado é aquele que possui um salário muito acima daquilo que ele entrega em quadra. Eles são pagos mais do que o "valor de mercado" típico para jogadores de habilidades e realizações semelhantes.

 Muitas vezes esses atletas superfaturados existem por conta de contratos longos que refletiam uma promessa de desempenho, jogadores que já foram muito bons e hoje desempenham pior mas mantêm seus altos salários (Algo bastante evidente pelos resultados) ou por conta do fator fora das quadras que é bem forte atualmente, como marketing e merchandising dos atletas.

Os super salários estão se tornando algo muito mais comum e isso é possivel observar pois maior parte dos atletas superfaturados jogaram nas temporadas mais recentes, em contraste aos atletas do inicio do século. Claro que ser algo superfaturado tem um pouco de subjetividade mas se olharmos apenas pelo o que é produzido dentro das quadras e esquecendo outros fatores, podemos ver que existem dezenas de salários que estão muito acima da média e isso demonstra a atual situação da NBA e da mídia.

Os chamados jogadores 'overpaids' aqui da nossa análise serão determinados por temporada, isto é, independente do jogador ser uma estrela histórica da NBA, se naquele ano em especifíco, ele ganhando um alto salário, performar mal dentro da quadra, muito provavelmente ele entrará na lista dos jogadores superfaturados naquela temporada. Iremos escolher os jogadores que tiverem uma maior diferença entre o salário real naquela temporada e o seu valor estimado de acordo com o modelo.
